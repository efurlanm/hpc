### openmpi

```python
%%bash
module load mpi/openmpi-x86_64
d=$(pwd)/lo
cd fftw-3.3.9
./configure CC=gcc --enable-mpi MPICC=mpicc --prefix=$d && \
make && \
make install
```

## mpi4py-fft

Install the library in a new enviroment

```python
%%bash
conda create --prefix ~/envs    # create a new environment
conda activate --stack ~/envs   # activate
conda install -c conda-forge mpi4py-fft    # install the library
conda install ipykernel         # the Notebook needs a new kernel
ipython kernel install --user --name=envs  # creates a new kernel
```
